```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlp
import seaborn as sb
%matplotlib inline
```

```python
df=pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.isnull().mean()
```

```python
sb.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='plasma')
```

```python
df[['gk_handling','gk_positioning','defending_marking','release_clause_eur','nation_jersey_number']].isnull()
```

Above heatmap and above table shows that following columns have more number of null values

```python
mlp.scatter(df['overall'],df['wage_eur'])
mlp.xlabel("Overall")
mlp.ylabel('Wages')
mlp.show()
```

Above this implies that the players whose overall attribute and skills are good are getting high wages


```python
mlp.bar(df['international_reputation'],df['wage_eur'],color='blue',width=0.4)
mlp.xlabel('Overall Attribute')
mlp.ylabel('Wages of Player')
mlp.show()
```

It shows that internataional reputation of a good player varies linearly with their wages/salary.Ti obvious that player with good reputation will earn more salary

```python
sb.histplot(data=df,x='international_reputation',hue='preferred_foot')
```

```python
df.weight_kg.plot.density(color='green')
mlp.title('Density Plot for weight')
mlp.show()
df.height_cm.plot.density(color='blue')
mlp.title('Density Plot for height')
mlp.show()
```

Above both plots are density plots of players weight and height respectively.
1. Averge weight of players is around 75-80 Kgs.
2. Average height of players is around 175-185 cms.
